# Merging & Concatentaing Data

```python
import pandas as pd
```

```python
bios = pd.read_csv("./data/bios.csv")
nocs = pd.read_csv("./data/noc_regions.csv")
results = pd.read_csv("./data/results.csv")
```

```python
bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left')
```

```python
bios_new.head()
```

```python
bios_new.rename(columns={"region": "born_country_full"}, inplace=True)
```

```python
bios_new.head()
```

```python
bios_new[bios_new["NOC_x"]!= bios_new["born_country_full"]][["name", "NOC_x", "born_country_full"]]
```

```python
ind = bios[bios['born_country'] == 'IND'].copy()
usa = bios[bios['born_country'] == 'USA'].copy()
```

```python
ind[['name', 'born_city']]
```

```python
new_df = pd.concat((ind, usa))
new_df.head()
```

```python
combined_df = pd.merge(results, bios, on='athlete_id', how='left')
```

# Handling NULL Values

```python
coffee = pd.read_csv("./warmup-data/coffee.csv")
coffee.head()
```

```python
import numpy as np
coffee.loc[[0,1], 'Units Sold'] = np.nan
```

```python
coffee.head()
```

```python
coffee.info()
```

```python
coffee.fillna(coffee['Units Sold'].mean()) # interpolate()
```

```python
coffee['Units Sold'] = coffee['Units Sold'].interpolate()
```

```python
coffee.head()
```

```python
coffee.loc[[0,1], 'Units Sold'] = np.nan
coffee.dropna()
```

```python
coffee[coffee['Units Sold'].isna()] # notna()
```

# Aggregating data

```python
bios.head()
```

```python
bios['born_country'].value_counts()
```

```python
bios[bios['born_country'] == 'IND']['born_city'].value_counts()
```

```python
coffee.groupby(['Coffee Type'])['Units Sold'].sum()
```

```python
coffee['Price'] = np.where(coffee['Coffee Type'] == 'Espresso', 4.99, 3.99)
coffee['Revenue'] = coffee['Units Sold'] * coffee['Price']
```

```python
coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum', 'Price': 'mean'})
```

```python
coffee
```

```python
pivot = coffee.pivot(columns='Coffee Type', index='Day', values = 'Revenue')
```

```python
pivot
```

```python
bios['born_date'] = pd.to_datetime(bios['born_date'])
bios.groupby(bios['born_date'].dt.year)['name'].count().reset_index().sort_values('name', ascending=False)
```

# Advanced Functionality

```python
coffee.head()
```

```python
coffee['yesterday_revenue'] = coffee['Revenue'].shift(2)
```

```python
coffee['pct_change'] = coffee['Revenue'] / coffee['yesterday_revenue'] * 100
```

```python
coffee.head(10)
```

```python
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
```

```python
bios.sort_values(['height_rank']).sample(10)[['name', 'height_rank']]
```

```python
coffee['cummulative_revenue'] = coffee['Revenue'].cumsum()

coffee.head(10)
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
