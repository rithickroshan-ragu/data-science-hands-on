# Adding/Removing Columns

```python
print(coffee.head(5))
coffee["Price"]=4.99
print(coffee.head(5))
```

```python
import numpy as np
coffee["new_price"] = np.where(coffee["Coffee Type"]=="Espresso", 5.99, 4.49)
coffee.head(5)
```

```python
# Drop rows
coffee.drop(0, inplace=True)
coffee.head(5)
```

```python
coffee.drop(columns=["Price"], inplace=True)
coffee.head(5)
```

```python
coffee_new = coffee
print(id(coffee_new))
print(id(coffee))
```

```python
coffee_new = coffee.copy()
print(id(coffee_new))
print(id(coffee))
```

```python
coffee["revenue"] = coffee["Units Sold"] * coffee["new_price"]
coffee.head(5)
```

```python
coffee.rename(columns={"new_price":"price"}, inplace = True)
coffee.head(3)
```

```python
bios_new = bios.copy()
bios_new["first_name"] = bios_new["name"].str.split(" ").str[0]

bios_new.head(3)
```

```python
bios_new.info()
```

```python
bios_new["born_datetime"] = pd.to_datetime(bios_new["born_date"])

bios_new.head(3)
```

```python
bios_new["born_year"] = bios_new["born_datetime"].dt.year

bios_new[["name", "born_year"]].head(3)
```

```python
bios_new.to_csv("./data/bios_new.csv", index=False)
```
